## Task 3: Relationship between Film Revenue and Budget

```python
sns.scatterplot(df['revenue'],df['budget'])
plt.title("Relationship between revenue and film budget")
plt.show()
```

```python
df.plot(x='budget',y='revenue',style='*', color='maroon')
plt.title("Relationship between film revenue and film budget")
plt.xlabel('budget')
plt.ylabel('revenue')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr())
plt.show()
```

```python
rel = df[['revenue','budget']].corr
sns.heatmap(rel())
plt.show()
```

## Task 4: Distribution of Languages in Film

```python
df['original_language'].unique()
sns.scatterplot(df['original_language'],df['revenue'],color='purple')
plt.show()
```

```python
sns.scatterplot(df['original_language'],np.log1p(df['revenue']),color='orange')
plt.show()
```

```python
sns.boxplot(df['original_language'],df['revenue'])
plt.show()
```

```python
df[df['original_language']==df['original_language'].max()]
```

```python
plt.subplots(figsize=(10,10))
df.groupby(['original_language'])['revenue'].sum().sort_values(ascending = False).head(20).plot(kind='pie')
plt.title("Distribution of languages in films", fontsize=24)
plt.xlabel('original_language', fontsize=24)
plt.ylabel('revenue', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```
